```python
import rstoolbox as rs
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = rs.io.parse_rosetta_fragments('rstoolbox/tests/data/wauto.200.3mers.gz')
df.head()
```

```python
df.shape
```

```python
xx = df[(df['frame'] <= 10) & (df['neighbor'] <= 100)]
```

```python
xx = df[(df['frame'] > 20) & (df['frame'] <= 30) & (df['neighbor'] <= 100)]
```

```python
def add_fragments( self, fragments, ini, how='replace' ):
    frags = fragments.copy()
    print(frags['frame'].unique())
    df = self.copy()
    columns = ['frame', 'neighbor', 'position']
    if how.lower() == 'replace':
        df = df.add_fragments(fragments, ini, 'replace')
    if how.lower() == 'append':
        frags['frame'] = frags['frame'] + ini - frags['frame'].values[0]
        print(frags['frame'].unique())
        print(df.shape)
        df = pd.concat([df, frags]).sort_values(columns).reset_index(drop=True)
        print(df.shape)
        df = df.coerce()
    return df
    
```

```python
df2 = add_fragments(df.copy(), xx.copy(), 10, 'replace')
```

```python
print(df2['frame'].unique())
print(df2['neighbors'].unique())
```

```python
fig = plt.figure(figsize=(25, 10))
ax0 = plt.subplot2grid((2, 1), (0, 0))
ax0.bar(range(len(df2.groupby('frame'))), [max(y['neighbor']) for x, y in df2.groupby('frame')])
ax1 = plt.subplot2grid((2, 1), (1, 0))
ax1.bar(range(len(df2.groupby('frame'))), [y['neighbors'].unique()[0] for x, y in df2.groupby('frame')])

plt.tight_layout()
plt.show()
```
